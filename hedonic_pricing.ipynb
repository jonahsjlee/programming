```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

df = pd.read_csv('./sample_data/cars_hw.csv')
df.head()
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split


# Split the data into training (80%) and test (20%) sets
train_data, test_data = train_test_split(df, test_size=0.2, random_state=42)

# Verify the split
print(f"Training set size: {len(train_data)}")
print(f"Test set size: {len(test_data)}")
```

```python
from sklearn.linear_model import LinearRegression

# Initialize models for different predictors
model_x1 = LinearRegression()
model_x2 = LinearRegression()
model_x3 = LinearRegression()

# Define predictor and target variables
X_train_x1 = train_data[['Mileage_Run']]  # Replace 'x1' with the actual column name
X_train_x2 = train_data[['Seating_Capacity']]
X_train_x3 = train_data[['Make_Year']]
y_train = train_data['Price']   # Replace 'target' with the actual target column name

# Fit models
model_x1.fit(X_train_x1, y_train)
model_x2.fit(X_train_x2, y_train)
model_x3.fit(X_train_x3, y_train)
```

```python
from sklearn.metrics import mean_squared_error, r2_score
import numpy as np

# Function to calculate RMSE and R^2
def evaluate_model(model, X_train, X_test, y_train, y_test):
    # Predictions
    y_train_pred = model.predict(X_train)
    y_test_pred = model.predict(X_test)

    # RMSE and R^2
    rmse_train = np.sqrt(mean_squared_error(y_train, y_train_pred))
    rmse_test = np.sqrt(mean_squared_error(y_test, y_test_pred))
    r2_train = r2_score(y_train, y_train_pred)
    r2_test = r2_score(y_test, y_test_pred)

    return rmse_train, rmse_test, r2_train, r2_test

# Define test sets
X_test_x1 = test_data[['Mileage_Run']]
X_test_x2 = test_data[['Seating_Capacity']]
X_test_x3 = test_data[['Make_Year']]
y_test = test_data['Price']

# Evaluate each model
results_x1 = evaluate_model(model_x1, X_train_x1, X_test_x1, y_train, y_test)
results_x2 = evaluate_model(model_x2, X_train_x2, X_test_x2, y_train, y_test)
results_x3 = evaluate_model(model_x3, X_train_x3, X_test_x3, y_train, y_test)

# Display results
print("Model x1 - RMSE (Train):", results_x1[0], "RMSE (Test):", results_x1[1], "R^2 (Train):", results_x1[2], "R^2 (Test):", results_x1[3])
print("Model x2 - RMSE (Train):", results_x2[0], "RMSE (Test):", results_x2[1], "R^2 (Train):", results_x2[2], "R^2 (Test):", results_x2[3])
print("Model x3 - RMSE (Train):", results_x3[0], "RMSE (Test):", results_x3[1], "R^2 (Train):", results_x3[2], "R^2 (Test):", results_x3[3])
```

I learned that overfitting occurs when a model performs well on a training set but poorly on the test set. Simple linear models with no interactions are less likely to overfit than more complex models.
