Q2, #4

```python
def clean_sentence_lengths(df):
    sentence_lengths = pd.to_numeric(df['ImposedSentenceAllChargeInContactEvent'], errors='coerce')

    initial_missing_count = sentence_lengths.isnull().sum()
    print(f'Initial missing values: {initial_missing_count}\n')

    sentence_lengths[df['SentenceTypeAllChargesAtConvictionInContactEvent'] == 4] = 0
    # Sentence type 9 treated as NaN
    sentence_lengths[df['SentenceTypeAllChargesAtConvictionInContactEvent'] == 9] = np.nan

    final_missing_count = sentence_lengths.isnull().sum()
    print(f'Final missing values after cleaning: {final_missing_count}\n')

    return sentence_lengths

df['ImposedSentenceAllChargeInContactEvent'] = clean_sentence_lengths(df)

print(pd.crosstab(df['ImposedSentenceAllChargeInContactEvent'].isnull(),
                  df['SentenceTypeAllChargesAtConvictionInContactEvent']))
```

I changed the sentences with 4 to be 0, and 9 to be nan
